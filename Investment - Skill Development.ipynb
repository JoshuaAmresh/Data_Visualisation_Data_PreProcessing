#### Data Cleaning

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age',data=train,palette='winter')
```

```python
#from the above plot, we see that average age of class 1 is 37,class 2 is 29,class 3 is 24.
#can use this method to impute the nulls
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        
        elif Pclass ==2:
            return 29
        
        else:
            return 24
        
    else :
        return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

#### Ordinal Encoding

```python
#types
#1. label encoding
#2. target guided ordinal encoding
```

```python
#label encoding, take same salary prediction by degree
#sort the degrees by (be,phd,bsc) rank and give high numerical value to the most weight category
# like phd - 1, be - 2, bsc - 3
```

#### Target Guided Ordinal Encoding

```python
# consider a feature X1 has categories a,b,c,d. we will take X1 and also the target variable (say classification prob) 0 and 1
# for all the instances of a, we will calculate the mean of target variable
# similar for all categories. and based on which category has the higher mean, we will arrange rank
# say means of a= 0.6, b=0.9, c=0.5 --> we will give label encoding as b=1,a=2,c=3 by ranking based on higher mean
```

#### Mean Encoding

```python
# similar to target guided ordinal encoding, but instead of giving labels, we will just give the mean itself to the variable
# so from the above cell example, the output of mean encoding will be, a=0.6, b=0.9, c=0.5 (not 2 or 3 or 1)
# applicable in places like pincode categorical variables
```

```python
# this function will remove the first feature that is correlated with any other feature

def correlation(dataset,threshold):
    col_corr = set() # set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i,j]) > threshold: #we are interested in absolute coeff  values
                colname = corr_matrix.columns[i] #getting the name of the column
                col_corr.add(colname)
    return col_corr
```

##### Using Z-score
Formula Z = (X - Mu)/sigma(standard dev)

```python
outliers = []
def detect_outliers(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outliers_pt = detect_outliers(dataset)
```

```python
outliers_pt
```

##### Using Inter-Quantile Range
75% - 25% values in the dataset

Steps

1. Arrange data in increasing order
2. Calculate 1st and 3rd quantile (q1 and q3)
3. Find interquantile range (q3-q1)
4. Find lower bound q1-(1.5*iqr)
5. Find upper bound q3+(1.5*iqr)

Anything that lies outside of lower and upper bound is an outlier

```python
np.sort(dataset)
```

```python
q1,q3 = np.percentile(dataset,[25,75])
```

```python
q1,q3
```

```python
iqr = q3-q1
iqr
```

```python
lower_bound = q1-(1.5*iqr)
upper_bound = q3+(1.5*iqr)
```
